```python
import numpy as ny
import pandas as pd
from sklearn.model_selection import train_test_split
from  sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('heart_disease_data.csv')
data.head(5)
```

```python
X=data.drop(columns='target',axis=1)
Y=data['target']
```

```python
import warnings
warnings.simplefilter('ignore')
```

# Spliting the Data into Training and Testing the Model

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

# Modeling the Data

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_predict=model.predict(x_train)
train_acc=accuracy_score(x_predict,y_train)
```

```python
print("Accuracy Score for Training Data:",train_acc)
```

```python
x_predict=model.predict(x_test)
test_acc=accuracy_score(x_predict,y_test)
```

```python
print("Accuracy Score for Testing Data:",test_acc)
```
